```python
import pandas as pd
import matplotlib.pyplot as plt

# Carica dataset
df = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/alcohol-consumption/drinks.csv")

# 1. Visualizza i primi 10 paesi ordinati per total_litres_of_pure_alcohol
print(df.sort_values(by='total_litres_of_pure_alcohol', ascending=False).head(10))

# 2. Media birra, vino e distillati
print("Media birra:", df['beer_servings'].mean())
print("Media vino:", df['wine_servings'].mean())
print("Media distillati:", df['spirit_servings'].mean())

# 3. Colonna alcohol_index
df['alcohol_index'] = (df['beer_servings'] + df['wine_servings'] + df['spirit_servings']) / 3

# 4. Paese con valore massimo
print("Massimo alcohol_index:", df.loc[df['alcohol_index'].idxmax()])

# 5. Paesi con pi di 100 birre all'anno
print(df[df['beer_servings'] > 100])

# 6. Bar chart dei primi 10
df_top10 = df.sort_values(by='total_litres_of_pure_alcohol', ascending=False).head(10)
df_top10.plot(kind='bar', x='country', y='total_litres_of_pure_alcohol')
plt.title('Top 10 consumo di alcol')
plt.show()

# 7. Line plot con wine_servings
df_sorted = df.sort_values(by='wine_servings')
plt.plot(df_sorted['country'], df_sorted['wine_servings'])
plt.xticks(rotation=90)
plt.title("Wine servings per paese")
plt.show()
```

```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset("lukebourne/data_jobs")
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

# 1. stipendio medio
print(df['salary_year_avg'].mean())

# 2. job count per paese
print(df['location'].value_counts())

# 3. intervallo stipendi
print("Min:", df['salary_year_avg'].min())
print("Max:", df['salary_year_avg'].max())

# 4. Bar plot stipendi medi per tipo lavoro
df.groupby('job_title_short')['salary_year_avg'].mean().sort_values(ascending=False).plot(kind='barh')
plt.xlabel('Stipendio medio')
plt.title('Stipendio medio per tipo lavoro')
plt.show()
```
